```python
ipl_df = pd.read_csv("ipl.csv")
ipl_df.head()
```

```python
columns_removel = ["mid","venue","batsman","bowler","striker","non-striker"]
# columns_removel
```

```python
ipl_df.drop(labels =  columns_removel, axis = 1, inplace = True)
```

```python
ipl_df["bat_team"].unique()
```

```python
present_team = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians', 'Kings XI Punjab','Royal Challengers Bangalore', 'Delhi Daredevils','Sunrisers Hyderabad']
```

```python
ipl_df = ipl_df[(ipl_df["bat_team"].isin(present_team)) & (ipl_df["bowl_team"].isin(present_team))]
```

```python
ipl_df
```

```python
print(ipl_df["bat_team"].unique().shape)
print(ipl_df["bowl_team"].unique().shape)
```

```python
ipl_df["overs"]>=5.0
```

```python
ipl_df = ipl_df[ipl_df["overs"]>=5.0]
ipl_df
```

```python
from datetime import datetime
ipl_df["date"] = ipl_df["date"].apply(lambda x: datetime.strptime(x , "%Y-%m-%d"))
```

```python
ipl_df.info()
```

```python
encoded_df = pd.get_dummies(data = ipl_df , columns= ["bat_team","bowl_team"])
encoded_df
```
